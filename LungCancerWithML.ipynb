```python
#Data-ların yüklənməsi
import pandas as pd
data=pd.read_csv('lung-cancer.csv')
data
```

```python
#Təkrar data-ların göstərilməsi
data[data.duplicated()]
```

```python
#Təkrar data-ların silinməsi və indekslərin yenidən verilməsi
data.drop_duplicates(inplace=True)
data.reset_index(inplace=True,drop=True)
data
```

```python
#Outlier-ların silinməsi və indekslərin yenidən verilməsi
import scipy.stats as stats
data['Z_score_value']=stats.zscore(data['AGE'])
drop_index = data[(data['Z_score_value']>3) | (data['Z_score_value']<-3)].index
data.drop(drop_index, inplace = True)
data.drop(['Z_score_value'], axis = 1, inplace= True)
data.reset_index(inplace=True,drop=True)
data
```

```python
#NULL dəyərlərin sayının göstərilməsi
data.isna().sum()
```

```python
#Data haqqında ətraflı məlumatların verilməsi
data.info()
data.describe()
```

```python
#Cinsiyyətə görə sayların çıxarılması
data['GENDER'].value_counts()
```

```python
#Cinsiyyətə uyğun pasta qrafikinin göstərilməsi
import matplotlib.pyplot as plt
plt.pie(data['GENDER'].value_counts(), labels=data['GENDER'].unique(), autopct='%1.1f%%')
plt.show()
```

```python
#Yaş və xəstəlik arasındakı əlaqənin qrafiklə göstərilməsi
plt.scatter(data['AGE'],data['LUNG_CANCER'])
```

```python
#Cinsiyyət üçün M və F-in müvafiq olaraq 1 və 2 ilə əvəzlənməsi
k = 1
for i in data.GENDER.unique():
  data['GENDER'].replace(i, k, inplace=True)
  k+=1
data
```

```python
#Ağciyər xərçəngi üçün YES və NO-nun müvafiq olaraq 1 və 2 ilə əvəzlənməsi
k = 1
for i in data.LUNG_CANCER.unique():
  data['LUNG_CANCER'].replace(i, k, inplace=True)
  k+=1
data
```

```python
#Test və train üçün data-ların hazırlanması
X=data.drop(columns=['LUNG_CANCER'],axis=1)
y=data['LUNG_CANCER']
```

```python
#Test və train data-lara parçalama
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100)
```

```python
#Support Vector Classifier alqoritminin tətbiqi və qiymətləndirilməsi
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
clf = SVC(C = 10, kernel='linear')
clf.fit(X_train,y_train)
y_pred = clf.predict(X_test)
print(accuracy_score(y_test, y_pred)*100)
print(classification_report(y_test, y_pred))
```

```python
#Gradient Boosting Classifier alqoritminin tətbiqi və qiymətləndirilməsi
from sklearn.ensemble import GradientBoostingClassifier
clf = GradientBoostingClassifier(random_state=50)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print(accuracy_score(y_test, y_pred)*100)
print(classification_report(y_test, y_pred))
```

```python
#Gaussian Naïve Bayes alqoritminin tətbiqi və qiymətləndirilməsi
from sklearn.naive_bayes import GaussianNB
clf = GaussianNB()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print(accuracy_score(y_test, y_pred)*100)
print(classification_report(y_test, y_pred))
```

```python
#Random Forest Classifier alqoritminin tətbiqi və qiymətləndirilməsi
from sklearn.ensemble import RandomForestClassifier
clf=RandomForestClassifier(random_state=50)
clf.fit(X_train, y_train)
y_pred=clf.predict(X_test)
print(accuracy_score(y_test, y_pred)*100)
print(classification_report(y_test, y_pred))
```

```python
#Decision Tree Classifier alqoritminin tətbiqi və qiymətləndirilməsi
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(max_depth=10, random_state=123)
clf.fit(X_train,y_train)
y_pred = clf.predict(X_test)
print(accuracy_score(y_test, y_pred)*100)
print(classification_report(y_test, y_pred))
```
